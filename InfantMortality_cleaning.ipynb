```python
# Creating function to combine races belonging to other.
def combine_other(row):
    return row['Infant Deaths (Asian/PI)'] + row['Infant Deaths (Am. Ind.)'] + row ['Infant Deaths (Multi-Race)']
nc_infant['Infant Deaths(Other)'] = nc_infant.apply(combine_other, axis=1)

# Dropping the original columns
nc_infant.drop(columns=['Infant Deaths (Asian/PI)', 'Infant Deaths (Am. Ind.)', 'Infant Deaths (Multi-Race)'], inplace=True)
```

```python
# Removing * From the dataframe 
def remove_asterisks(value):
    if isinstance(value, str) and '*' in value:
        return value.replace('*', '')
    return value

# Applying function to the entire df
nc_infant = nc_infant.applymap(remove_asterisks)
```

```python
nc_infant.head(50)
```

```python
# Converting some columns to numeric
def convert_columns_to_numeric(df, exceptions):
    for column in df.columns:
        if column not in exceptions:
            df[column] = pd.to_numeric(df[column], errors='coerce')
    return df

# Converting columns except County of Residence and total C.I.
exceptions = ['County of Residence', 'Total C.I.']
nc_infant = convert_columns_to_numeric(nc_infant, exceptions)
```

```python
nc_infant.info()
```

```python
# Doing the same thing for rates 
# Creating function to combine races belonging to other.
def combine_other2(row):
    return row['Infant Death Rate (Asian/PI)'] + row['Infant Death Rate (Am.Ind.)'] + row ['Infant Death Rate (Multi-Race)']
nc_infant['Infant Death Rate(Other)'] = nc_infant.apply(combine_other2, axis=1)

# Dropping the original columns
nc_infant.drop(columns=['Infant Death Rate (Asian/PI)', 'Infant Death Rate (Am.Ind.)', 'Infant Death Rate (Multi-Race)'], inplace=True)
```
